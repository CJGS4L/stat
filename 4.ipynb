```python
import numpy as np
import math
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.distributions.empirical_distribution import ECDF
import scipy.stats as ss

#1.1#

list_1 = []
list_2 = []
list_3 = []
list_4 = []

for i in range(1000):
      list_1.append(sum(np.random.binomial(n=1, p=0.5, size=20)))
      list_2.append(sum(np.random.uniform(0,1, size=20)))
      list_3.append(sum(np.random.exponential(1, size=20)))
      list_4.append(sum(np.random.standard_cauchy(size=20)))

print(list_4)

x_1 = np.arange (np.mean(list_1)-3*math.sqrt(np.var(list_1)), np.mean(list_1)+3*math.sqrt(np.var(list_1)), 0.001)
ecdf=ECDF(list_1)
plt.step(ecdf.x, ecdf.y)
plt.plot(x_1,ss.norm.cdf(x_1,np.mean(list_1),math.sqrt(np.var(list_1))))
plt.show()

x_2 = np.arange (np.mean(list_2)-3*math.sqrt(np.var(list_2)), np.mean(list_2)+3*math.sqrt(np.var(list_2)), 0.001)
ecdf=ECDF(list_2)
plt.step(ecdf.x, ecdf.y)
plt.plot(x_2,ss.norm.cdf(x_2,np.mean(list_2),math.sqrt(np.var(list_2))))
plt.show()


x_3 = np.arange (np.mean(list_3)-3*math.sqrt(np.var(list_3)), np.mean(list_3)+3*math.sqrt(np.var(list_3)), 0.001)
ecdf=ECDF(list_3)
plt.step(ecdf.x, ecdf.y)
plt.plot(x_3,ss.norm.cdf(x_3,np.mean(list_3),math.sqrt(np.var(list_3))))
plt.show()

x_4 = np.arange (np.mean(list_4)-3*math.sqrt(np.var(list_4)), np.mean(list_4)+3*math.sqrt(np.var(list_4)), 0.001)
ecdf=ECDF(list_4)
plt.step(ecdf.x, ecdf.y)
plt.plot(x_4,ss.norm.cdf(x_4,np.mean(list_4),math.sqrt(np.var(list_4))))
plt.show()













```
