```python
import numpy as np
import math
import scipy.stats
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss

#1#
list_prob = np.random.uniform(0,1,size=1000)
list_exp = []


for i in range(1000):
  list_exp.append(-math.log(1-list_prob[i]))
plt.hist(list_exp, density = True, alpha = 0.7)
plt.hist(np.random.exponential(1,size=1000), density = True, alpha = 0.7)
plt.title('Exp(1)')
plt.show()

list_cauchy = []
for i in range(1000):
  list_cauchy.append(math.tan(math.pi*(list_prob[i]-0.5)))

plt.hist(list_cauchy, density = True, alpha = 0.7)
plt.hist(np.random.standard_cauchy(size=1000), density = True, alpha = 0.7)
plt.title('Cauchy')


```

```python
#2#
R=1
x_1 = []
x_2 = []
for i in range(1000):
  u = np.random.uniform(-R,R)
  v = np.random.uniform(-R,R)
  if (u*u + v*v <= R):
    x_1.append(u)
    x_2.append(v)
plt.scatter(x_1,x_2)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()



x = []
y = []
R = 1
for i in range(10000):
	theta = np.random.uniform(0,2*math.pi)
	r = np.random.uniform(0,R)
	x.append(math.sqrt(r*R)*math.cos(theta))
	y.append(math.sqrt(r*R)*math.sin(theta))

plt.scatter(x,y, s = 1)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()
```

```python
#3#
list_exp1 = []
for i in range(1000):
  y = np.random.binomial(1, 0.5)
  if y == 1:
    list_exp1.append(np.random.exponential(1))
  else:
    list_exp1.append((-1)*np.random.exponential(1))
plt.hist(list_exp1, density = True)
plt.show()

list_exp = []
for i in range(1000):
  y = np.random.binomial(1, 0.5)
  if y == 1:
    list_exp.append(np.random.exponential(1))
  else:
    list_exp.append((-1)*np.random.exponential(1/2))

plt.hist(list_exp, density = True)
```

```python
#5#
list_tr = []
f = lambda x: (1-x)/2  if -1 < x < 0 else (1+x)/2
for i in range(10000):
  y = np.random.uniform(-1,1)
  if np.random.binomial(n=1, p = f(y)) == 0 :
    list_tr.append(y)
plt.hist(list_tr, density = True, bins = 100)
plt.xlim(-2, 2)
plt.show()
```
