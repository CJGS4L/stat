```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
import statistics as st
from statsmodels.distributions.empirical_distribution import ECDF

#1.1#
n = 100
X = np.random.normal(loc = 0, scale = 1, size =n)
x = np.arange(-3, 3, 0.01)
F = ss.norm.cdf(x, loc = 0, scale = 1)
ecdf=ECDF(X)
plt.plot(x, F)
plt.plot(ecdf.x, ecdf.y)
plt.show()


X = np.random.uniform(0, 1, n)
x = np.arange(-1, 2, 0.01)
F = ss.uniform.cdf(x,0,1)
ecdf=ECDF(X)
plt.plot(x, F)
plt.plot(ecdf.x, ecdf.y)
plt.show()

n = 10000
m = 10
p = 1/2
X = np.random.binomial(m, p, n)
x = np.arange(-1, m+1, 0.01)
F = ss.binom.cdf(x, m, p)
ecdf=ECDF(X)
plt.plot(x, F)
plt.plot(ecdf.x, ecdf.y)
plt.show()

#1.2#
n=10
X=np.random.normal(0, 1, size = n)
#x=np.arange(-3,3,0.01)
#F = st.norm.cdf(x,loc=0,scale=1)
ecdf=ECDF(X)
#plt.plot(x, F)
y1=(1.36/math.sqrt(n))
y2=math.sqrt(-math.log(0.05/2)/(2*n))
plt.plot(ecdf.x, ecdf.y)
plt.plot(ecdf.x, ss.norm.cdf(ecdf.x)+y1)
plt.plot(ecdf.x, ss.norm.cdf(ecdf.x)-y1)
```

```python

#2.1#
n = 50
X = [0]*n
Y = np.random.normal(loc = -3, scale = math.sqrt(8), size = n)
for i in range(n):
  X[i] = (Y[i]+3)/math.sqrt(8)
print(ss.kstest(X, ss.norm.cdf))

Pvalue = []
k = 1000
for j in range(k):
  for i in range(n):
    Y = np.random.normal(loc = -3, scale = math.sqrt(8), size = n)
    X[i] = (Y[i]+3)/math.sqrt(8)
  Pvalue.append(ss.kstest(X, ss.norm.cdf).pvalue)

ax=np.arange(1,k+1)/k
print(Pvalue)
plt.plot(np.sort(Pvalue),ax,drawstyle='steps-post')
plt.plot(ax,ax)
plt.show()

#2.2#
n = 100
m = 100
X = [0]*n
Y = np.random.binomial(m, 1/2, size = n)
for i in range(n):
  X[i] = (Y[i]-m/2)/math.sqrt(m*(1/4))

Pvalue = []
for j in range(k):
  for i in range(n):
    Y = np.random.binomial(m, 1/2)
    X[i] = (Y-m/2)/math.sqrt(m*(1/4))
  Pvalue.append(ss.kstest(X, ss.norm.cdf).pvalue)
ax=np.arange(1,k+1)/k
print(Pvalue)
plt.plot(np.sort(Pvalue),ax,drawstyle='steps-post')
plt.plot(ax,ax)
plt.show()

#2.3#
n = 30
m = 5
X = [0]*n
for i in range(n):
  Y = np.random.uniform(0, 1, size = m)
  X[i] = (sum(Y)-m/2)/math.sqrt(m/12)
print(ss.kstest(X, ss.norm.cdf))

Pvalue = []

for j in range(k):
  for i in range(n):
    Y = np.random.uniform(0, 1, size = m)
    X[i] = (sum(Y)-m/2)/math.sqrt(m/12)
  Pvalue.append(ss.kstest(X, ss.norm.cdf).pvalue)
ax=np.arange(1,k+1)/k
print(Pvalue)
plt.plot(np.sort(Pvalue),ax,drawstyle='steps-post')
plt.plot(ax,ax)
plt.show()
```
