```python
#3#
print(data.groupby('Пол').mean(numeric_only=True)[['Сумма.баллов']])

print(data.groupby('Аттестат.с.отличием.оценка').mean(numeric_only=True)[['Сумма.баллов']])

print(data.groupby(['Пол','Аттестат.с.отличием.оценка']).mean(numeric_only=True)[['Сумма.баллов']])

print(data[(data['ЕГЭ...матем.'] < 70)].shape[0])
print(data[(data['ЕГЭ...матем.'] > 70) & (data['ЕГЭ...матем.'] <= 80)].shape[0])
print(data[(data['ЕГЭ...матем.'] > 80) & (data['ЕГЭ...матем.'] <= 90)].shape[0])
print(data[(data['ЕГЭ...матем.'] > 90)].shape[0])


data['Диапазон.мат.'] = pd.cut(x=data['ЕГЭ...матем.'], bins = [0, 70, 80, 90, 100], labels=['<70', '70-80', '80-90', '90+'])
data['Диапазон.русс.'] = pd.cut(x=data['ЕГЭ...рус.яз.'], bins = [0, 70, 80, 90, 100], labels=['<70', '70-80', '80-90', '90+'])
data1 = data.groupby(['Диапазон.мат.','Диапазон.русс.'])['Сумма.баллов'].agg('count').reset_index()
data1.rename(columns = {'Сумма.баллов':'Количество'})
```
