```python
import plotly as plt
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
from scipy. stats import norm

x = np.arange(-20, 20, 0.01)

fig = go.Figure()

for i in range (1,8):
  fig.add_trace(go.Scatter(x=x, y = norm. pdf (x,0,i)))

fig.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = [i for i in range(1000)]
rval1 = np.random.binomial(n=100,p=0.04,size=1000)
rval2 = np.random.poisson(lam=4,size=1000)
rval3 = np.random.binomial(n=1000,p=0.004,size=1000)
rval4 = np.random.binomial(n=10,p=0.4,size=1000)

plt.figure()
plt.subplot(2,3,1)
plt.hist(rval1, bins=range(15))
plt.title('Binomial(100,0.04)')
plt.show()

plt.subplot(2,3,2)
plt.hist(rval2, bins=range(15))
plt.title('Poiss(4)')
plt.show()

plt.subplot(2,3,3)
plt.hist(rval1, bins=range(15), alpha=0.5)
plt.hist(rval2, bins=range(15), alpha=0.5)
plt.title('Poiss(4) and Binomial(100,0.04)')
plt.show()

plt.subplot(2,3,4)
plt.hist(rval3, bins=range(15))
plt.title('Binomial(1000,0.004)')
plt.show()

plt.subplot(2,3,5)
plt.hist(rval4, bins=range(15))
plt.title('Binomial(10,0.4)')
plt.show()

plt.subplot(2,3,6)
plt.hist(rval3, bins=range(15), alpha=0.5)
plt.hist(rval4, bins=range(15), alpha=0.5)
plt.title('Binomial(1000,0.004) and Binomial(10,0.4)')
plt.show()
```
